```python
import os
import duckdb
import bso_coverage_tools as bct
from dotenv import load_dotenv
```

```python
# Load server environment
load_dotenv(os.getcwd() + '\\.env')
```

```python
def bso_apply_alex_coverage(row):
    res = bct.analyse_from_doi(row["doi"], cli=False)
    return res

def bso_apply_doi_coverage(row):
    res = bct.analyse_from_openalex_work(row, cli=False)
    return res

def bso_task_coverage(n_rows=100, from_doi=True):
    # print(f"[BSO] Start applying coverage with n={n_rows}")
    with duckdb.connect(f"md:?motherduck_token={os.getenv('MD_TOKEN')}") as con:
        df = con.sql(f"SELECT * FROM openalex_2021 WHERE doi IS NOT NULL ORDER BY id LIMIT {n_rows}").df()
        # print(f"[BSO] Start analyze of {df.shape[0]} entries")
        res_df = df.apply(bso_apply_doi_coverage, axis=1) if from_doi==True else df.apply(bso_apply_alex_coverage, axis=1)
        display(res_df)
```

```python
%timeit bso_task_coverage(n_rows=10, from_doi=True)
```

```python
%timeit bso_task_coverage(n_rows=10, from_doi=False)
```
